```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img =cv2.imread("C:\\Users\\aedpu\\OneDrive\\Documents\\Guhan.M\\III Year\\Tom Cruise\\xray.jpeg")
```

```python
img
```

```python
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
```

```python
plt.imshow(img)
```

```python
sobelX = cv2.Sobel(img,cv2.CV_64F,1,0)
sobelY = cv2.Sobel(img,cv2.CV_64F,0,1)
```

```python
sobelX = np.uint8(np.absolute(sobelX))
sobelY = np.uint8(np.absolute(sobelY))
```

```python
combind = cv2.bitwise_or(sobelX,sobelY)
```

```python
plt.imshow(sobelY)
```

```python
lab = cv2.Laplacian(img, cv2.CV_64F,ksize=3)
lab = np.uint8(np.absolute(lab))
```

```python
combined_edges = cv2.addWeighted(combind, 0.5, lab, 0.5, 0)
```

```python
titles = ['image ','sobelY','combind','Laplacian','combindedges']
image = [img,sobelX,sobelY,combind,lab,combined_edges]

for i in range(5):
    plt.subplot(2,3,i+1),plt.imshow(image[i])
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```
